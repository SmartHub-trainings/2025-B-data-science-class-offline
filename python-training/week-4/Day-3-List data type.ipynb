```python
import random
```

```python
ll=3
ul=20
max_trial=5
magic_number = random.randint(ll,ul)
for i in range(max_trial):
    trials_left = max_trial-(i+1)
    user_input=int(input(f"Make a guess({ll}-{ul}): "))
    if user_input==magic_number:
        print("Great!!\nYou got it right!")
        break
    else:
        print("Please, use your brain and try again.")
        if trials_left==0:
            print("You're a failure!!")
        else:
            print(f"You have {trials_left} trial(s) left")
```

Q 62: A particular retailer is having a 60 percent off sale on a variety of discontinued
products. The retailer would like to help its customers determine the reduced price
of the merchandise by having a printed discount table on the shelf that shows the original prices and the prices after the discount has been applied. Write a program that
uses a loop to generate this table, showing the original price, the discount amount,
and the new price for purchases of $4.95, $9.95, $14.95, $19.95 and $24.95. Ensure
that the discount amounts and the new prices are rounded to 2 decimal places when
they are displayed.

```python
prices =[4.95,9.95,14.95,19.95,24.95]
discount_per=0.6
for price in prices:
    discount = round(price*discount_per,2)
    new_price = round(price-discount,2)
    print(discount)
    print(new_price)
    print()
    
```

```python
prices =[4.95,9.95,14.95,19.95,24.95]
discount_per=0.6
print("price\tdiscount\tNew price")
for price in prices:
    discount = round(price*discount_per,2)
    new_price = round(price-discount,2)
    print(f"{price}\t{discount}\t\t{new_price}")

    
    
    
```
