## Break and Continue Statements

```python
for i in range(10):
    if i ==5:
        continue
    print(i)
```

```python
for i in range(10):
    if i ==5:
        break
    print(i)
```

```python
for i in range(10):
    if i ==5 or i ==7 or i ==9:
        continue
    print(i)
```

```python
for i in range(10):
    if i ==5:
        continue
        print("Now, i have to skip")
    print(i)
```

```python
for i in range(10):
    if i ==5:
        print("Now, i have to skip")
        continue
    print(i)
```

```python
for i in range(5):
    name = input("Enter your name: ")
    print(f"Hello, {name}")
```

```python
for i in range(5):
    name = input("Enter your name: ")
    if len(name)!=5:
        continue
    print(f"Hello, {name}")
```
