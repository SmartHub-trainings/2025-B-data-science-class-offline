```python
try:
    a= float(input("Enter a number: "))
    b= float(input("Enter another number: "))
    c=a/b
except ZeroDivisionError :
    print("You can not divide by 0")
except ValueError:
    print("Both entries must be numbers")
except:
    print("You did something wrong!!")
else:
    print(c)
finally:
    print("Done!!")
```

Q 144: Create a program that adds line numbers to a file. The name of the input file will be
read from the user, as will the name of the new file that your program will create.
Each line in the output file should begin with the line number, followed by a colon
and a space, followed by the line from the input file.

```python
input_filename=input("Enter the name of file to read from: ")
output_filename=input("Enter the name of file to read to: ")
input_file_handler=open(input_filename)
content= input_file_handler.readlines()
print(content)
```

```python
try:
    input_filename=input("Enter the name of file to read from: ")
    output_filename=input("Enter the name of file to read to: ")
    input_file_handler=open(input_filename)
except FileNotFoundError:
    print("The requested file does not exist.")
else:
    content= input_file_handler.readlines()
    print(content)
    line_number =1
    for line in content:
        new_line =f"{line_number}: {line}"
        line_number = line_number +1
        print(new_line)
```

```python
try:
    input_filename=input("Enter the name of file to read from: ")
    output_filename=input("Enter the name of file to read to: ")
    input_file_handler=open(input_filename)
except FileNotFoundError:
    print("The requested file does not exist.")
else:
    content= input_file_handler.readlines()
    print(content)
    number_of_items =len(content)
    for i in range(number_of_items):
        new_line =f"{i+1}: {content[i]}"
        print(new_line)
```
