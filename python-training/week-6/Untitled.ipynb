```python
def shift_char(char,shift_key=3):
    is_lower=char.islower()
    alphabets='abcdefghijklmnopqrstuvwxyz'
    char_index=alphabets.index(char.lower())
    shift_char_index=char_index+shift_key
    if shift_char_index>25:
        shift_char_index= shift_char_index % 26
    shifted_char = alphabets[shift_char_index]
    if is_lower:
        return shifted_char
    else:
        return shifted_char.upper()
```

```python
def encode_text(text,shift_key=3):
    encrypted_chars=[]
    for letter in text:
        if letter.isalpha():
            encrypted_chars.append(shift_char(letter,shift_key))
        else:
            encrypted_chars.append(letter)
    cypher_text="".join(encrypted_chars)
    return cypher_text
```
